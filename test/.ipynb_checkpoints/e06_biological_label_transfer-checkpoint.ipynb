```python
import os
import sys

# 添加 AlignCell 到 sys.path，允许 Python 从该路径加载模块
sys.path.append('/Gao2/dzh/trace_embryo/new/new/AlignCell')

from data_processing.data_loader import DataLoader
from data_processing.split_cell import split_cell
from data_processing.preprocess import Preprocess_gene
from data_processing.data import TwitterDataset

%config InlineBackend.figure_format = 'svg'
```

```python
#输出基因表达矩阵，要求行名为基因名称，列名为细胞名称
import pandas as pd
import numpy as np

new_expression = pd.read_csv('./example/heart_count_dat.txt',sep="\t",index_col=0)
new_expression
```

```python
#添加样品信息表，要求至少含有cell以及cell_type两列的data.frame
fil_type=pd.read_csv('./example/heart_type.txt',sep="\t")
fil_type
```

```python
#按照type的8：2比例划分query和target，除训练模型外，实际应用过程中可以不需要此步骤
train_data, test_data, train_expression_matrix, test_expression_matrix = split_cell(fil_type, new_expression, 
                                                                                    group_column='type', cell_name_column='cell',
                                                                                    random_state=9)
```

```python
#将target数据信息，给target添加唯一标签
loader1 = DataLoader(expression_matrix_1=train_expression_matrix, cell_labels_1=train_data, top_genes=1000, mode="target",cell_name_column='cell')

# 处理数据
ref_dat, ref_label, ref_cell_labels = loader1.process_data()

print("更新后的 ref_cell_labels:")
print(ref_cell_labels)
```

```python
#将query数据信息，给query添加唯一标签
loader1 = DataLoader(expression_matrix_1=test_expression_matrix, cell_labels_1=test_data, top_genes=1000, mode="query",cell_name_column='cell')

# 处理数据
que_dat, que_label, que_cell_labels = loader1.process_data()

print("更新后的 que_cell_labels:")
print(que_cell_labels)
```

```python
#读取构建target信息向量，其中w2v路径为默认的模型训练好的，如果有使用自己训练的w2v向量可以自行添加，注意版本号
preprocess = Preprocess_gene(ref_dat, sen_len=1000)
embedding = preprocess.make_embedding()
gene_name = preprocess.sentence_word2idx()
gene_label = preprocess.labels_to_tensor(ref_label)
```

```python
#读取构建query信息向量，其中w2v路径为默认的模型训练好的，如果有使用自己训练的w2v向量可以自行添加，注意版本号
preprocess2 = Preprocess_gene(que_dat, sen_len=1000)
embedding2 = preprocess2.make_embedding()
gene_name2 = preprocess2.sentence_word2idx()
gene_label2 = preprocess2.labels_to_tensor(que_label)
```

```python
#mode有两种模式，当训练模型时使用'positive_negative'模式，构建三元损失函数样品对，当应用训练完的模型预测时mode选择'single'模式
ref_dataset = TwitterDataset(x=gene_name, z=gene_label, mode="single")
que_dataset = TwitterDataset(x=gene_name2, z=gene_label2, mode="single")
```

```python
from get_result.Get_emb import AlignCell_Emb
from AlignCell import TransformerEncoder, PositionalEncoding, EncoderLayer, FeedForward
```

```python
#读取已训过好的model
emb_dat=AlignCell_Emb("test_heart/FineTurn_no_mtDNA.model")
```

```python
#获取target和query中最相近的向量，并进行标签迁移，要求是ref_dataset，que_dataset顺序
result = emb_dat.evaluate(ref_dataset, que_dataset)
```

```python
target_mapping = dict(zip(ref_cell_labels['cell_id'], ref_cell_labels[['cell', 'type']].values))

# 2. 在 df_query 中添加新的列 'target_cell' 和 'target_cell_types'，并将映射的值赋给它们
result['target_cell'] = result['test_label'].map(lambda x: target_mapping.get(x, [None, None])[0])
result['target_cell_types'] = result['test_label'].map(lambda x: target_mapping.get(x, [None, None])[1])

# 3. 查看结果


test_cell_labels_new = pd.merge(que_cell_labels, result[['test_label','rel_label', 'target_cell', 'target_cell_types']], 
                    left_on='cell_id', right_on='rel_label', how='left')

test_cell_labels_new['result'] = (test_cell_labels_new['type'] == test_cell_labels_new['target_cell_types'])
# 4. 查看结果
print(test_cell_labels_new)
```
